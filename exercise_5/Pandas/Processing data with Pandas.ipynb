## <font color=#ff6600>Exploring Some Basics Data Operations</h3></font>

```python
# import pandas module as pd
import pandas as pd

# define filepath to variable (fp)
# find what type of file it is
# Read the data

fp = r"C:\Users\Gustavocolmenares\Documents\SchoolStaff\GIS_PCC\Courses_Training\Geo_python_Course\Exercise-5\Pandas\Kumpula-June-2016-w-metadata.txt"
# bc is a *.txt file check file & see what's the separeted format; in this case 'comma-separated format'
df = pd.read_csv(fp, sep=',', skiprows= 8) # we used skiprows=8 because the first 8 lines contained metadata
df.head()
```

## <font color=green>Calculations with dataframes</h3></font>

```python
# define a new column "DIFF" 
df['DIFF'] = 0.0 # column will contain float data
#print dataframe
df.head()
```

```python
# Check data types
df.dtypes
```

```python
# calculates the max and min differences and update the 'DIFF' column with the result
df['DIFF'] = df['MAX'] - df['MIN']

# print dataframe head to check if the calculation was completed
df.head()
```

```python
# Create columns ON-THE-FLY and update its cells with a calculation
df['DIFF_Min'] = df['TEMP']- df['MIN']
#print dataframe updated with the new column "TEMP_Min"
df.head()
```

## <font color=green> You can do any calculation that you need using math algebra</h3></font>

```python
# Create a new column called 'TEMP_Celsius and calculate the temperature to Celsius
df['TEMP_Celsius'] = (df['TEMP']-32)/(9/5)

#print updated dataframe with new column 'TEMP_Celsius'
df.head()
```

## <font color=green>Selecting data using indices</h3></font>

```python
# <variable> = dataFrame[start_index:stop_index, {intervals}].
rows5 = df[0:5]
rows5
```

```python
# dataframe.loc[start_index:stop_index, list_of_columns] returns the values of an individual row from specific position as a pd.Series

row8 = df.loc[8] # create variable to containt the selected column
print(row8, '\n')

# To access the value of an individual cell by selecting its index

print(row8['TEMP'],'\n') # from the selected row, print the value of specific 

# You can also select multiple rows simultaniously 
temp_5to10 = df.loc[5:10, 'TEMP'] # variable will contain the value from the selected rows and the selected column

print(temp_5to10)
```

```python
# Selecting multiple columns values between a fixed or given index frame

temp_5to10 = df.loc[5:10, ['TEMP', 'TEMP_Celsius']]
print(temp_5to10)
```

```python
temps_only  = df[['TEMP', 'TEMP_Celsius']]
print(temps_only)
```


## <font color=purple>Filtering and Updating Data</h3></font>

```python
# Using the .loc[] we can select a specific column and or rows based on a condition statement 

w_temps = df.loc[df['TEMP_Celsius']> 15] # Select all the temperatures above 15 degree from 'TEMP_Celsius' column
w_temps
```

```python
# Combining multiple criteria can be done using (&) AND operator or (|) OR operator.
# Combine multiple criterias to select temperatures above 15 degrees that were recorder half of june in 2016

w_temp2 = df.loc[(df['TEMP_Celsius']> 15) & (df['YEARMODA'] >= 20160615)]
w_temp2
```

```python
w_temp3 = df.loc[(df['TEMP']>= 60) & (df['YEARMODA'] == 20160626)]
w_temp3
```

```python
w_temp4 = df.loc[(df['TEMP_Celsius'] >18) & (df['DIFF_Min']>= 5)]
w_temp4
```

```python
# Index values on left shows index from original DataFrame. 
# To reset index use: reset_index(drop =True)
# By default reset_index() would make a new column called (index) that keep track of prevous index

# Reset index
w_temp2 = w_temp2.reset_index(drop=True)
w_temp2
```

## <font color=#408000> Dealing with Missing Data</h3></font>

```python
# Let's change the value of some cells [0:4] from column [TEMP_celsius] to 'NaN' to show missing data 

w_temp2.loc[:4, 'TEMP_Celsius'] = None
print(w_temp2)
```

### Now that we have some missing data  
 > How to deal with <b>MISSING DATA</b>:
     1. Remove them from DataFrame
     2. Fill them with some value

```python
# Option 1: Remove the rows that contain missing data from the DataFrame 
    # Use dropna(subset=<'column_name'> or <column_index>) function

w_temp_clean = w_temp2.dropna(subset=['TEMP_Celsius'])
print(w_temp_clean)
```

```python
# Option 2: Fill the NoData with some value using fillna(<value_to_fill_with>) function.
# Let's fill it with the value 0

w_temp2_na_filled = w_temp2.fillna(0)
w_temp2_na_filled
```

## <font color=#cc4400>Sorting the Data</h3></font>

> To sort your data (descending/ascending) base on values in some column.
> Use from Pandas; sort_values(by=<'column_name'>)

```python
# Sorting the values in ascending order based on 'TEMP' column
sorted_temp_a = df.sort_values(by='TEMP')
sorted_temp_a
```

```python
# Now lets sort the 'TEMP' column in descending order 

sorted_temp_d = df.sort_values(by='TEMP', ascending=False)
sorted_temp_d
```

## <font color=#cc9900>Rounding and Finding Unique Values</h3></font>
<p><b>R</b>ounding is easy by using round() function.</p>

```python
# lets round the values from TEMP_Celsius with 0 -decimals

# Create a new column that contains rounded numbers
df['Celsius_Rounded'] = df['TEMP_Celsius'].round(0)
df
```

<p><b>E</b>xtract Unique values from a column by using <i>"unique_values()"</i> function.</p> 




```python
# Lets get unique Celsius values

unique = df['Celsius_Rounded'].unique() # unique variable will contain unique Celsius values
print(unique) # its returns an array of unique values from that column
type(unique) # print the data type of unique variable
```

```python
# Sometimes a long array list values doesn't show all of the values.
# for this we can print the array list as a list using function list(<variable>)
# if you want to create a list from the array list do:
lst_unique = list(unique) # create a new variable 
lst_unique
```

```python
# Lets find how many days with unique temperature did we have in June 2016

uniq_temp_days = len(unique) # create variable that contain the lenght of unique array list
print('There was {} days with unique mean temperature in June 2016'.format(uniq_temp_days))
```

## <font color=#cc9900>Writing Data... </h3></font>
<p>To be able to write data that you have analyzed is very easy using Pandas.
<br>Check for more info here: <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html">DataFrame to CSV or Excel</a></p> 


```python
# Use to_csv to save the DataFrame in a csv file.

# Create a variable that contain the name of the [New CSV] file.
output_df = 'Kumpula_temps_June_2016.csv'

# Save the DataFrame to csv file using: <dataframe>.to_csv(<path_of_New_file>,sep=',' )
df.to_csv(output_df, sep=",")
```

```python
# lets create another csv file, but this time:
# Omiting index column use: (index=False)
# Only with 1 decimal floating value use: float_format="%.1f" - changing value of 1 for 2 would write 2 decimals, etc

output_df2 = 'Kumpula_temps_above15_June_2016.csv' # variable with name of NEW csv file
# Use: dataframe.to_csv(<path_of_file>, sep=',', index= False, float_format= %1f)
df.to_csv(output_df2, sep=',', index=False, float_format="%.1f") 
```
